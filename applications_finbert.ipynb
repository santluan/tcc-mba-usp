# Aplicações do algoritmo FinBERT para Análise de Sentimentos

## FinBERT

```python
from transformers import pipeline

pipe = pipeline("text-classification", model="ProsusAI/finbert")

print(pipe("Tesco share price jumps as Q3 sales top estimates"))
```

```python
import pandas as pd

df = pd.read_csv("financial_sentences.csv")

df['sentiment_finbert'] = df['Sentence'].apply(lambda x: pipe(x)[0]['label'])

df.head()
```

```python
from sklearn.metrics import classification_report

print(classification_report(df['Sentiment'], df['sentiment_finbert']))
```

```python
df.to_csv("financial_news_with_finbert_sentiment.csv", index=False)
```
